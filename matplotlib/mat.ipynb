```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
```

```python
y=2*x
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=4)
plt.title("x vs y")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

```python
x
```

```python
y1=2*y
y1
```

```python
plt.plot(x,y,color='g',linestyle=':')
plt.plot(x,y1,color='r',linewidth=3)
plt.title('two lines in a plot')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color="g",linestyle=':')
plt.subplot(2,2,2)
plt.plot(x,y1,color="r",linewidth=3)
plt.title("2 plot in one table")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(True)
plt.show()
```

```python
student={"ishu":88,"chinkey":33,"sandeep":00}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.title("distribution or marks of students")
plt.xlabel("students names")
plt.ylabel("marks")
plt.show()
```
