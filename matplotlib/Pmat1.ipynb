```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,color="r",linestyle=":",linewidth=4)
plt.title("this is x vs y ")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

```python
y1=2*y
```

```python
y1
```

```python
plt.plot(x,y,color='g',linestyle=':')
plt.plot(x,y1,color='r',linewidth=3)
plt.title('two lines in a plot')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color="g",linestyle=":")
plt.title("two plot in one")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(True)
plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("two plot in one")
plt.xlabel("this is x axis")
plt.ylabel("this is y1 axis")
plt.grid(True)
plt.show()
```

```python
student={"sandeep":22,"pooja":20,"hemlata":18}
name=list(student.keys())
age=list(student.values())
```

```python
plt.bar(name,age)
plt.show()
```

```python
plt.barh(name,age,color="g")
plt.title("this is age chart")
plt.xlabel("students names")
plt.ylabel("students age")
plt.show()
```
