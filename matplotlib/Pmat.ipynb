```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
y=2*x
```

```python
plt.plot(x,y,color="g",linestyle=":",linewidth=4)
plt.show()
```

```python
y1=2*y
```

```python
y1
```

```python
plt.plot(x,y,color="g",linestyle=":")
plt.plot(x,y1,color="r",linewidth=5)
plt.title("two plot in one")
plt.xlabel("this is x axis")
plt.ylabel("this y axis")
plt.grid(True)
plt.show()

```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color="g",linestyle=":")
plt.title("two in one")
plt.xlabel("THIS IS X ASIC")
plt.ylabel("ths is y lable")
plt.grid(True)
plt.subplot(2,2,2)
plt.plot(x,y1,color="r",linewidth=4)
plt.title("two in one")
plt.xlabel("THIS IS X ASIC")
plt.ylabel("ths is y lable")
plt.grid(True)
plt.show()
```

```python
student={"rahul":25,"sandeep":24,"himanshu":45,"kapil":15}
names=list(student.keys())
age=list(student.values())
```

```python
plt.bar(names,age)
plt.title("haha")
plt.xlabel("student names")
plt.ylabel("students marks")
plt.show()
```

```python
plt.barh(names,age,color="g")
plt.title("haha")
plt.xlabel("students marks")
plt.ylabel("students names")
plt.show()
```

```python
a=[10,20,30,50,44,66,33]
b=[2,4,5,6,7,6,7]
plt.scatter(a,b,marker="*",color="green",s=100)
plt.title("this is scatter plot")
plt.xlabel("this is x axis")
plt.ylabel("this is y label")
plt.show()
```

```python
c=[1,2,3,4,6,8,9]
```

```python
plt.scatter(a,b,c="g")
plt.scatter(a,c,c="r",marker="*")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(a,b)
plt.subplot(2,2,2)
plt.scatter(a,c,color="g",marker="*",s=100)
plt.show()
```

```python
#pie chart
fruit=["apple","mango","grapes","banana"]
quantity=[10,20,30,40]
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
fruit=["apple","mango","grapes","banana"]
quantity=[10,20,30,40]
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','green','red','blue'])
plt.show()
```

```python
#doughnut_chart
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
